```python
import threading
import time
```

```python
def take_orders():
    for i in range(1, 4):
        print(f"Taking order for #{i}")
        time.sleep(1)
```

```python
def brew_chai():
    for i in range(1, 4):
        print(f"Brewing chai for order #{i}")
        time.sleep(2)
```

# Creating Threads

```python
order_thread = threading.Thread(target=take_orders)
brew_thread = threading.Thread(target=brew_chai)
```

```python
order_thread.start()
brew_thread.start()
```

# Waiting for threads to finish

```python
order_thread.join()
brew_thread.join()
```

```python
print(f"All Ordres taken and chai brewed")
```
