```python
import threading
import time
```

```python
def boil_milk():
              print(f"Boiling Milk...")
              time.sleep(2)
              print(f"Done Boiling Milk")
```

```python
def toast_bun():
    print(f"Toasting bun...")
    time.sleep(3)
    print(f"Done Toasting bun")
```

```python
start = time.time()

t1 = threading.Thread(target=boil_milk)
t2 = threading.Thread(target=toast_bun)

t1.start()
t2.start()

t1.join()
t2.join()

end = time.time()
print(f"Breakfast is ready in {end - start:.2f}")
```
