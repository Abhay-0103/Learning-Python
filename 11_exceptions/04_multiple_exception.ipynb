```python
def process_order(item, quantity):
    try:
        price = {"Apple": 50}[item]
        cost = price * quantity
        print(f"The Total cost is {cost}")
    except KeyError:
        print("Sorry lowde we dont have this in menu")
    except TypeError:
        print("Lowde Quantity should be a number")

process_order("Apple", 2)
process_order("Banana", "meow")
```
