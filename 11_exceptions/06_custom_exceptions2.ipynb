```python
class OutOfIngredientsError(Exception):
    pass

def make_briyani(masala, briyaniPowder):
    if masala == 0 or briyaniPowder == 0:
        raise OutOfIngredientsError("Cannot make briyani, ingredients are missing!")
    print("Making delicious briyani with masala and briyani powder!")

make_briyani(1, 1)  # This will work
make_briyani(0, 1)  # This will raise OutOfIngredientsError
```
