```python
class InvalidBriyaniError(Exception): pass

def bill(briyani_type, plants):
    menu = {
        "chicken": 200,
        "mutton": 300
    }

    try:
        if briyani_type not in menu:
            raise InvalidBriyaniError("That Briyani is not available Gandu....")
        if not isinstance(plants, int):
            raise TypeError("Number of briyani must be an integer, you idiot...")
        total = menu[briyani_type] * plants
        print(f"Appka Bill  {plants} plants ka {briyani_type} briyani: {total} rupees")
    except Exception as e:
        print("Error: ", e)
    finally:
        print("Thank you for visiting our Briyani shop, you idiot...")

bill("mmix", 2)  # Invalid briyani type
bill("chicken", "two")  # Invalid number of plants
bill("mutton", 3)  # Valid order
```
