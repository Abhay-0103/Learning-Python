# Old Way

```python
file = open("order.txt", "w")
try:
    file.write("Meoww Ghopp Ghopp - 2 baar")
finally:
    file.close()
```

# Mordern Way 

```python
with open("order.txt", "w") as file:
    file.write("Meoww Ghopp Ghopp - 2 baar")
```

- This way is better because it automatically takes care of closing the file after the block of code is executed, even if an error occurs. It ensures that resources are properly managed and reduces the chances of leaving files open unintentionally.
