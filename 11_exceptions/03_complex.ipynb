```python
def serve_chai(flavor):
    try:
        print(f"Preparing {flavor} chai...")
        if flavor == "unknown":
            raise ValueError("Sorry, Maga nam atara e  flavor illa !")
    except ValueError as e:
            print("Error: ", e)
    else: 
         print(f"{flavor} chai is ready to serve!")
    finally:
         print("Next Shade Plesae")

serve_chai("masala")
serve_chai("unknown")
```
